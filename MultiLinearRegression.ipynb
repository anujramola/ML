```python
import pandas as pd
```

```python
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset.info
```

```python
dataset.head(15)

```

```python
state = dataset['State']
from sklearn.preprocessing import LabelEncoder
state_le = LabelEncoder()
state_le_final = state_le.fit_transform(state)
state_le_final
```

```python
state_le_final.shape
```

```python
state_le_final_2D = state_le_final.reshape(-1,1)
```

```python
from sklearn.preprocessing import OneHotEncoder
state_ohe = OneHotEncoder()
state_prefinal = state_ohe.fit_transform(state_le_final_2D)
state_final = state_prefinal.toarray()
```

```python
state_final
```

```python
state_dummytrap_final = state_final[ : , 0:2 ]
```

```python
state_dummytrap_final
```

```python
dataset.head(2)
```

```python
y = dataset['Profit']
```

```python
X = dataset[['R&D Spend','Administration','Marketing Spend']]
```

```python
import numpy 
X_final = numpy.hstack((X ,state_dummytrap_final ))
X_final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
from sklearn.model_selection import train_test_split
X_final_train, X_final_test, y_train, y_test = train_test_split(X_final, y, test_size=0.20, random_state=42)
```

```python
X_final_train
```

```python
y_train
```

```python
model.fit(X_final_train , y_train)
```
